Burada Doğum analizi  yapıp verıyı grafiklerle görselleştireceğiz.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore');
```

```python
df=pd.read_csv('births.csv')
```

```python
# Çubuk grafik
plt.bar(df['gender'], df['births'])
plt.xlabel('Gender')
plt.ylabel('births')
plt.title('Gender Distribution')
plt.show()
```

```python
# Yıllara göre doğumları grupla
births_by_year = df.groupby('year').size()

# Çizgi grafiği oluştur
plt.plot(births_by_year.index, births_by_year.values)
plt.xlabel('Yıl')
plt.ylabel('Doğum Sayısı')
plt.title('Yıllara Göre Doğum Sayısı')
plt.show()
```

```python
# Histogram
df['births'].plot(kind='hist', bins=30)
plt.xlabel('Births')
plt.ylabel('Frequency')
plt.title('Distribution of Births')
plt.show()
```

```python
# Çizgi grafiği
df.plot(x='year', y='births', kind='line')
plt.xlabel('Year')
plt.ylabel('Births')
plt.title('Births Over the Years')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Grafik boyutunu 10x6 olarak ayarlayın

df.boxplot(column='births', by='year')
plt.xlabel('Year')
plt.ylabel('Births')
plt.title('Births Distribution by Year')
plt.show()
```

```python
sns.countplot(x=df['year'],hue=df.gender)
plt.xticks(rotation=90);
```
